```python
plt.figure(figsize=(6,4))

sns.countplot(x='Class', data=df)

plt.title("Fraud vs Normal Transactions")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

df['Amount_Scaled'] = scaler.fit_transform(df[['Amount']])
```

```python
df[['Amount', 'Amount_Scaled']].head()
```

```python
fraud_cases = df[df['Class'] == 1]

fraud_cases.head()
```

```python
fraud_cases.shape
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='Class', y='Amount_Scaled', data=df)

plt.title("Fraud vs Normal Transaction Amounts")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Time'], bins=50)

plt.title("Transaction Time Distribution")

plt.show()
```

```python
df['Suspicious'] = df['Amount_Scaled'].apply(
    lambda x: 1 if x > 3 else 0
)
```

```python
df['Suspicious'].value_counts()
```

```python
suspicious_transactions = df[df['Suspicious'] == 1]

suspicious_transactions.head()
```

```python
df.to_csv("cleaned_fraud_data.csv", index=False)
```
